Creating 0-d tensors (i.e, scalars). ndim is a dimension of a vector.

```python
integer = torch.tensor(1234)
decimal = torch.tensor(3.14)

print(f"interger is a {integer.ndim}-dimensional Tensor: {integer}")
print(f"interger is a {integer.ndim}-dimensional Tensor with shape: {integer.shape}")
print(f"decimal is a {decimal.ndim}-dimensional Tensor: {decimal}")
```

Vectors and lists are 1-d tensors. shape is a 'size' of dimension.

```python
fibonacci = torch.tensor([1, 1, 2, 3, 5, 8])
count_to_100 = torch.tensor((range(100)))

count_to_100_arange = torch.arange(100) # alternative way to create tensor, immediately allocates memory

print(f"fibonacci is a {fibonacci.ndim}-dimensional Tensor with shape: {fibonacci.shape}")
print(f"count_to_100 is a {count_to_100.ndim}-dimensional Tensor with shape: {count_to_100.shape}")
print(f"count_to_100_arange is a {count_to_100_arange.ndim}-dimensional Tensor with shape: {count_to_100_arange.shape}")
```

Now I'll create 2-d (i.e, matrices) and higher-rank tensors. In image processing 4-d tensors are used with dimensions coressponding to: batch size, number of colors channels, image height, and image width.

```python
matrix = torch.tensor([[1, 2, 3], [4, 5, 6]])
matrix_zeros = torch.zeros(3, 4) # matrix with 3 rows and 4 columns with 0
matirx_ones = torch.ones(3,4)
matrix_rand = torch.rand(3,4)

print(f"matrix is a {matrix.ndim}-dimensional Tensor with shape: {matrix.shape}")

assert isinstance(matrix, torch.Tensor), "matrix is not a torch Tensor object"
assert matrix.ndim == 2, "matrix must be 2-d torch Tensor"

'''Defining a 4-d Tensor for images: size 10x3x256x256'''
# You can think of this as 10 images where each image is RGB 256 x 256.

images = torch.zeros(10, 3, 256, 256)

assert isinstance(images, torch.Tensor), "must be Tensor object"
assert images.ndim == 4, "must be 4-d"
assert images.shape == (10, 3, 256, 256), "incorrect shape"

print(f"images is a {images.ndim}-dimensional Tensor with shape: {images.shape}")
```

So the shape of a tensor provides the number of elements in each dimension. We can use slicing to access subtensors within higer-rank tensor. I'll use my previous matrix tensor.

```python
row_vector = matrix[1]
column_vector = matrix[:, 1]
scalar = matrix[0, 1]

print(f"'matrix': {matrix}")
print(f"`row_vector`: {row_vector}")
print(f"`column_vector`: {column_vector}")
print(f"`scalar`: {scalar}")
```

A convient way to think about and visualize computations in machine learning framework like PyTorch is in terms of graphs. We can define this graph in terms of tensors, which hold data, and the mathematical operations that act on these tensors in some order.

```python
# Creating the nodes in graph and initialize values
a = torch.tensor(15)
b = torch.tensor(61)

# Addition
c1 = torch.add(a, b)
c2 = a + b # PyTorch overrides the "+" operation so that it is able to act on Tensors.
print(f"c1: {c1}")
print(f"c2: {c2}")
```

Simple function doing some operations on tensors.

```python
# Defining Tensor Computation

# Construct simple computation function
def func(a, b):
  #changing integers to tensors
  a = torch.tensor(a)
  b = torch.tensor(b)

  c = torch.add(a, b)
  d = torch.subtract(b, 1)
  e = torch.multiply(c, d)
  return e
```

Calling this function

```python
# example value for a, b
a, b = 1.5, 2.5
# ecexute the computation
e_out = func(a, b)
print(f"e_out: {e_out}")
```
